```python
from PIL import Image
import numpy as np

np.random.seed(42)

def euclidean_distance(vx, vy):
   return np.linalg.norm(vx - vy)

def k_means_b(input_image, iterations):
    min_cost = float('inf')
    max_cost = float('-inf')
    min_cluster = []
    max_cluster = []
    min_cluster_means = []
    max_cluster_means = []
    
    for _ in range(iterations):
        cost = 0
        random_centroids = np.random.randint(0, 256, size=(3, 3))
        cluster_means = np.array(random_centroids)

        pixels = np.array(input_image).reshape((-1, 3))
        clusters = [[] for _ in range(3)]
        
        # Assign each pixel to the nearest cluster mean
        for pixel in pixels:
            distances = [euclidean_distance(pixel, mean) for mean in random_centroids]
            closest_cluster = np.argmin(distances)
            clusters[closest_cluster].append(pixel)

        for i in range(3):
            if clusters[i]:
                cluster_means[i] = np.mean(clusters[i], axis=0)

        # Calculate cost
        for closest_cluster, cluster_pixels in enumerate(clusters):
            cost += sum(euclidean_distance(pixel, cluster_means[closest_cluster]) for pixel in cluster_pixels)
                
        if cost < min_cost:
            min_cost = cost
            min_cluster = clusters
            min_cluster_means = cluster_means
            
        if cost > max_cost:
            max_cost = cost
            max_cluster = clusters
            max_cluster_means = cluster_means

    return min_cluster, max_cluster, min_cluster_means, max_cluster_means

def final_max(input_image, N):
    min_cluster, max_cluster, _, max_cluster_means = k_means_b(input_image, N)
    pixels = np.array(input_image).reshape((-1, 3))

    output_image = np.zeros_like(pixels)
    for i, cluster in enumerate(max_cluster):
        output_image[np.array(cluster)] = max_cluster_means[i]
    
    output_image = output_image.reshape(input_image.shape)
    
    return output_image

def final_min(input_image, N):
    min_cluster, _, min_cluster_means, _ = k_means_b(input_image, N)
    pixels = np.array(input_image).reshape((-1, 3))

    output_image = np.zeros_like(pixels)
    for i, cluster in enumerate(min_cluster):
        output_image[np.array(cluster)] = min_cluster_means[i]
    
    output_image = output_image.reshape(input_image.shape)
    
    return output_image



if __name__ == '__main__':
   # Load input images
   car_image = np.array(Image.open('car.png'))
   flower_image = np.array(Image.open('flower.png'))


   N = 30
   k = 3
   
   result_car_max = final_max(car_image, N)
   Image.fromarray(result_car_max.astype(np.uint8)).save("car_max_3.png")
   result_flower_max = final_max(flower_image, N)
   Image.fromarray(result_flower_max.astype(np.uint8)).save("flower_max_3.png")


   result_car_min = final_min(car_image, N)
   Image.fromarray(result_car_min.astype(np.uint8)).save("car_min_3.png")
   result_flower_min = final_min(flower_image, N)
   Image.fromarray(result_flower_min.astype(np.uint8)).save("flower_min_3.png")
```
